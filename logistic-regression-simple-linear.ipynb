## Logistic Regression simple linear 

```python
import pandas as pd #reading data
import numpy as np
import math 
from sklearn.linear_model import LogisticRegression, LinearRegression
import matplotlib.pyplot as plt #draw graph 
from matplotlib.pyplot import figure #draw graph
```

```python
filename = "hoursOfStudy.csv"
dataframe = pd.read_csv(filename)
dataframe.columns
# dataframe
```

```python
X = dataframe.drop(['Pass'], axis = 'columns')
X
y = dataframe.drop(['hours of study'], axis = 'columns')
y
```

```python
linear_regression = LinearRegression()
#train model 
linear_regression.fit(X,y)
```

```python
logistic_regression = LogisticRegression()
#train the model
logistic_regression.fit(X,y)
```

```python
#scatter: vẽ đồ thị từng điểm không nối vào nhaus
plt.xlabel('Hours of study')
plt.ylabel('Pass/Fail')
plt.ylim(-0.02 ,1.02)
plt.scatter(X, y, marker = '+', color = 'blue')
#plot (input, output)
plt.plot(X, linear_regression.coef_ *X + linear_regression.intercept_, color = 'green')
plt.plot(X, pd.DataFrame(logistic_regression.predict_proba(X))[1], color = 'purple')
#Fail
plt.plot(X, pd.DataFrame(logistic_regression.predict_proba(X))[0], color = 'red')
plt.scatter(3.1, logistic_regression.predict_proba([[3.1]])[0][1])
#biểu diễn điểm bất kì 
#Với linear regression thì cần Sigmoid, giá trị y của đường xanh thì sao khi qua sigmoid thì nó biến thành đường cong
# => y(output of linear regression) => sigmoid => 1/(1 + pow(e, -y))  
```

```python
#predict probability 
#xác xuất tại 3.1 
logistic_regression.predict_proba([[3.1]]) 
#2 giá trị là vì xác xuất thất bại là 0, và xác xuất thành công là 1
#0.39500117, 0.60499883 = 1
#xác xuất tại cả x 
logistic_regression.predict_proba(X) 
#biến nó thành dataframe cho dễ nhìn, 1 là xác xuất pass
pd.DataFrame(logistic_regression.predict_proba(X))[1]
logistic_regression.predict_proba([[3.1]])[0][1]
```
